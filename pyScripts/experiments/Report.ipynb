```python
def load_experiment(filename):
    experiment = pickle.load(open(filename,'rb'))

    ns=[]
    scores=[]
    values=[]
    for res in experiment['results']: 
        ns.append(res['output'][0])
        vals = res['output'][1]
        if max(vals)-min(vals) == 0:
            scores.append([0, 0, 0, 0])
        else:
            scores.append( 1 - (vals - min(vals)) / (max(vals) - min(vals)))
        values.append(vals)

    scores = np.array(scores)
    values = np.array(values)
    return ns, values, scores, experiment

def plot_results(experiment, values, x_trials=False):
    colors = [(0.867, 0.753, 0.082), (0.792, 0.075, 0.184), (0.337, 0.588, 0.788), (0.467, 0.714, 0.243)]
    plt.figure(figsize=(15,5))

    if x_trials:
        x_val = np.linspace(0, 1, len(experiment['trials']))
        plt.xticks(x_val, experiment['trials'])
    else:
        x_val = experiment['trials']
        plt.locator_params(nbins=20,axis='x')
    
    for i in range(4):
        if i == experiment['correct']:
            plt.plot(x_val, values[:,i], 
                     linewidth=3, color='black',label='Ans '+'abcd'[i])
        else:
            plt.plot(x_val, values[:,i], 
                     linewidth=1.5, color=colors[i],label='Ans '+'abcd'[i])    
    
    plt.title(experiment['expid']+':   '+experiment['comments'], y =-.2)
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.075),
                ncol=4, fancybox=True)
    
    ax = plt.axes()
    ax.set_frame_on(False)
    ax.tick_params(axis='y',width=0)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

    xmin, xmax = ax.get_xaxis().get_view_interval()
    ymin, ymax = ax.get_yaxis().get_view_interval()
    ax.add_artist(plt.Line2D((xmin, xmin), (ymin, ymax), color='black', linewidth=1))

    ax.grid(axis='y', which='major', linestyle='-',alpha=.4)
    ax.grid(axis='y', which='minor', linestyle='-',alpha=.2)
    ax.minorticks_on()
    ax.locator_params(nbins=20, axis='y')
    
    plt.ylim((ymin-.01,ymax+.1))
    plt.show()

def med_med_qa_distance(distances):
	return np.median(np.median(distances,axis=0))

def avg_avg_qa_distance(distances):
	return np.average(np.average(distances,axis=0))
```

# DrugBank Graph

DrugBank was exported to a connectivity list, which was then imported in [iGraph](http://igraph.org/) for Python, a network analysis package. We thus obtained a graph with 391089 edges and 1603540 edges (that can also be accessed in GraphML format). Every vertex corresponds to a DrugBank entity (via its 'uri' attribute) and every edge corresponds to any relation. Edges in the graph are assigned several weights:
- rel_weight: the number of relations between two entities
- deg_weight: the degree-corrected weight (Moore et al., 2011). The edge $(u,v)$ has weight $w(u,v) = \log(\deg(u)) + \log(\deg(v))$.
- inv_rel_weight: the inverse of the rel_weight such that nodes with many edges in between are 'close'.

The questions we used for this test experiment were extracted from a website ([pharmacology2000.com](http://www.pharmacology2000.com/)) that provides many in depth study questions on mainly pharmocology. We used this source mainly because it is digital, allowing easy access to (and filtering of) _all_ questions. Moreover it matched our dataset of choice: DrugBank. Some basic statistics on the question collection:

The question collection was filtered in several ways. Firstly, we only gathered multiple choice question sets with one correct answer and more than two options (two options often boiled down to a 'binary' true/false). This left us with 1360 questions with 3-5 options each. We then removed all questions where multiple answers where 'implicitely' true, so to speak. That is, questions of the form:
> [Question]
> 1. Some answer
> 2. Some other answer
> 3. Both 1. and 2.
> 4. None of the above.

After these restrictions, 654 questions remained with perhaps relatively little text: an average of 10.8 words in each question (sd 8.9) and an average of 3.9 words in each answer option (sd 2.9). Using a crude index, we ranked questions on text content and selected six high-ranked questions for which the four mc options all consisted of a drug (see below). This choice is completely motivated by our dataset (Drugbank). Using questions asking for some drugs at least excludes the possibility of 'misrepresenting' the possible answers in DrugBank. ## Representing the questions in the Drugbank Graph

We aim to rank the answers based on their distance to the question. So we first have to represent the question and the answers in the graph. As mentioned earlier, all of the multiple choice answers can (except for question 4) unambiguously be represented in Drugbank. The questions, however, rarely contain drugbank entities. But using a full-text search, many of the terms in the question, can be found in literals in drugbank. So we adopted this approach and performed a full-text search on drugbank for each of the boldfaced queries in the question text. We used Virtuoso's (sort of) built-in precision search and did _not_ further rank the search results. Consequently, the very first search result will generally not be a good representation of the term. Instead, we used the first $n$ results for each query to represent the question and analyzed the influence of the size of $n$.

Again, this is a very crude approach with many problems. For example, some queries might yield many results and could be 'overrepresented' when increasing $n$, while other possibly more relevant nodes could be 'underrepresented'. 

## Measuring the Q-A distance
After representing the question and answers, we are left with a list $Q=\{q_1, \dots, q_n\}$ of vertices for the question and a list $a=\{a_1, \dots, a_m\}$ of vertices for each of the answers $a,b,c,\dots$. For our six questions, the answers usually consist of one vertex. To rank the answers, we have to calculate some kind of distance to the question: the _qa-distance_. Starting point is the length of the shortest path $d(q,a)$ for each $a\in A$ and $q\in Q$. We can combine these to a distance $d(A,Q)$:

\begin{align}
d(A,Q) = \text{average}\bigl\{ \; \text{median}\{d(a,q): a \in A \}: \quad q\in Q\;\bigr\}
\end{align}

_**Note.** Instead of taking first the median, then the average, we could also take averages or medians only --- This perhaps deserves more thought. A comparison for one question suggests that median-median seperates the answers better._

The length of the shortest path depends on the edge_weight used. That is: either deg_weight, rel_weight, inv_rel_weight or no weight for the step distance.

## Scoring performance
The _performance_ or _score_ of each of the mc answers is measured by the relative, normalized distance to the question. More precisely, if $d_1, d_2, d_3$ and $d_4$ denote the distance from anwer 1, 2, 3 and 4 to the question respectively, if $M$ denotes the maximum and $m$ the minimum over these,  then the score of $d_i$ is

\begin{align}
s_i = 1 - \frac{d_i - m}{M-m}
\end{align}

In this way, $s_i = 1$ means that $d_i$ is closest to the question, i.e. the best guess. When $s_i = 0$ this is the 'worst guess'. The performance for the 'entire Question' is simply the performance of the correct answer.

### Influence of representation

These results suggest a way to improve the ranking procedure: look for the 'optimal' representation of the question and the answers.  
For example, take Question 5:

```python
ns,vals,scores,exp = load_experiment('Experiment 3/qa-experiment3-q5.p'); exp['comments'] = ''; plot_results(exp, scores, True)
```

When representing the question text by the first 50 to 100 search results, the distance between answer b, c, and d decreased. Apparently, the entities added made it more difficult to distinguish these three and where moreover 'closer' to answer a. That would clearly not be the best representation. Perhaps there is a way to improve this.
